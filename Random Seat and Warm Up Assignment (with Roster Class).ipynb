```python
# My classroom has groups of three. I have created a class called Roster which will 
# allow me to create a random seating chart and randomly choose a facilitator 
# for the student led warm up exercise.

import random
import math

class Roster():
    
    '''
    A roster contains a list of student names and a period number (integer).
    '''

    def __init__(self,names,period):
        self.names = names
        self.period = period
        
    def __str__(self):
        '''
        Returns class roster in alphabetical order
        '''
        alph = self.names.sort()
        namethem = ''
        for i in self.names:
            namethem += i
            namethem += '\n'
        return namethem     
    
    def shuffle(self):
        # shuffle roster
        random.shuffle(self.names)
        
        # check to see if someone is going to end up alone at a group and fix by
        # creating two groups of two instead
        if len(self.names)%3 == 1:
            self.names.insert(-2,'        ')
        
    def warmup(self):
        '''
        Picks random warm up facilitator and returns the string
        '''
        return(random.choice(self.names))
        
    def makepretty(self):
        '''
        Adds spaces to list items to make all items have a length of 8.
        Adds blank items to list for empty chair assignments.
        '''
        for i in range(len(self.names)):
            # shorten name to 8 characters if too long
            if len(self.names[i])>8:
                self.names[i]=self.names[i][0:8]
            # add spaces before and after names to center the display
            spaces = 8 - len(self.names[i])
            self.names[i] = (' '*math.floor(0.5*spaces)) + self.names[i] +\
            (' '*math.ceil(0.5*spaces))
        # add empty spaces to remaining seats with no students
        while len(self.names)<25:
            self.names.append('        ')
     
    def display_room(self):
        fac = self.warmup()
        self.makepretty()
        '''
        Prints out a text display of the seating chart and warm up facilitator.
        '''
        '\n'*10

        a = ' ________ ________        ________ ________        ________ ________'
        b = '|        |        |      |        |        |      |        |        |'
        c = '|________|________|      |________|________|      |________|________|'
        d = '     |        |               |        |               |        |'
        e = '     |________|               |________|               |________|'
        f = '              ________ ________        ________ ________'
        z = '             |        |        |      |        |        |'
        g = '             |________|________|      |________|________|'
        w = '                 |        |               |        |'
        h = '                 |________|               |________|'

        #Display Class Period
        print(f'\n                               PERIOD {self.period}')
        print(f'                          Facilitator: {fac.upper()}')

        #FRONT ROW
        print(a)
        print(b)
        print('|' + self.names[0] + '|' + self.names[1] + '|      |' + self.names[2] +\
             '|' + self.names[3] + '|      |' + self.names[4] + '|' + self.names[5] + '|')
        print(c)
        print(d)
        print('     |' + self.names[6] + '|               |' + self.names[7] + \
             '|               |' + self.names[8] + '|')
        print(e)
        print('\n')

        #MIDDLE ROW
        print (a)
        print (b)
        print('|' + self.names[9] + '|' + self.names[10] + '|      |' + self.names[12] +\
             '|' + self.names[13] + '|      |' + self.names[15] + '|' + self.names[16] + '|')
        print(c)
        print(d)
        print('     |' + self.names[11] + '|               |' + self.names[14] + \
             '|               |' + self.names[17] + '|')
        print(e)
        print('\n')

        #BACK ROW
        print(f)
        print(z)
        print('             |' + self.names[18] + '|' + self.names[19] + '|      |' + \
             self.names[21] + '|' + self.names[22] + '|')
        print(g)
        print(w)
        print('                 |' + self.names[20] +\
              '|               |' + self.names[23] + '|')
        print(h)


            
            
#UNSHUFFLED ROSTERS: eventually be able to import these from csv. Plug n chug for now
rawperiodone = ['Miniya','Taniya','Maya','Kayla','Ayiana','Simone']
rawperiodtwo = ['Arianna','Mya','Patrick','Eve','Christiana','Asha','Veda','Ernest',\
               'Andrew','Chyna','Marissa','Paul','Malachi','David','Jaden','Sean',\
               'Vicky','Andy O.','Taina','Nina']
rawperiodthree = ['Eliora','Darren','Victoria','Shayla','Kaya-Z','Armiya','Nikolai',\
                  'Allyson','Ciera','Sarah','Jayla','Nalah','Ashira','Meshaun','Karma',\
                 'Jahmiya','Sammy','Sabrina','Jade','Jayna','Flora','Aemilia','Kamai']
rawperiodfour = ['Kastenny','Amari','Arundhati','Dewan','Tamiyah','Amuzu','Kayla','Ruby',\
                'Raven','Jaida','Quinnyetta','Mason','Erik','Shaun']
rawperiodfive = ['Kwencie', 'Sydni', 'Ciera', 'MiAngel', "K'lah",'Taylor',\
                 'David','Kenia','Karl','Nyah', 'Samiya', 'Jay','Wend',\
                 'Brianna','Tim','Kamaya']
rawperiodsix = ['Jamie','Jamari','Quin','Kiona','Dayun','Joshvic','Luke',\
                'Elisha','Zaybian']


#Dictionary for period numbers and roster lists
periods = {1:rawperiodone,2:rawperiodtwo,3:rawperiodthree,4:rawperiodfour,\
          5:rawperiodfive,6:rawperiodsix}


#MAIN: user interface - asks for user to specify period 
#and returns seating chart with facilitator selected

while True:
    try:
        pd = int(input('Choose a period from 1 through 6:  '))
        if pd in range(1,7):
                break
        else:
            print("ERROR! You must input an integer from 1 - 6.")
    except ValueError:
        print('ERROR! You must input an integer from 1 - 6.')


ros = Roster(periods[pd],pd)
ros.shuffle()
ros.display_room()
```
